```python
def splitColsByParts(df,n):
    d = {}
    count=df.shape[1]
    parts=int(count/n)
    remains=int(count%n)
    fro=0
    to=0
    for one in range(1,n+1):
        if remains>0 and one==n:
            to=parts*one+remains
        else:
            to=parts*one
        
        '''print("from "+str(fro))
        print("to "+ str(to))'''
        d[one] =pd.DataFrame()
        d[one]=df.iloc[:,fro:to]
        fro=to
    return d
```
